```python
#Dfs 
def dfs(st,goal,graph,vis=None):
    if vis == None:
        vis=set()
    print(st)
    if st == goal:
        return 1
    vis.add(st)
    for child in graph[st]:
        if child not in vis:
            if dfs(child,goal,graph,vis):
                return True
            return False
```

```python
#Bfs
def bfs(st,goal,graph):
    vis=set()
    queue=[st]
    path=[]
    vis.add(st)
    while queue:
        node=queue.pop(0)
        path.append(node)
        if node == goal:
            return path
        for neighbours in graph[node]:
            if neighbours not in vis:
                vis.add(neighbours)
                queue.append(neighbours)
    return path
```
